```
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.despine()
```

```
df=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
dfg=df.groupby(by = 'WindDir9am').agg(['sum','mean'])
df.columns
```

```
sns.boxplot(x='WindDir9am',y='WindSpeed9am',hue='RainToday',data=df)
```

```
sns.jointplot(x='WindSpeed3pm',y='WindSpeed9am',data=df)
sns.despine()
```

```

sns.distplot(dfg['MaxTemp']['mean'])
sns.despine()
```

```
dfg['Rainfall']['mean'].plot(kind='barh')
sns.despine()
```

```
dfg['Evaporation']['sum'].plot(kind='bar')
sns.despine()
```

```
sns.jointplot(x='Sunshine',y='Evaporation',data=df)
sns.despine()
```

```
sns.violinplot(x='WindDir9am',y='MaxTemp',data=df)
sns.despine()
```

```
sns.violinplot(x='RainToday',y='WindSpeed9am',data=df)
sns.despine()
```

```
sns.boxplot(x='RainToday',y='Rainfall',data=df)
sns.despine()
```

```
sns.boxplot(x='RainToday',y='Temp9am',data=df)
sns.despine()
```

```
sns.countplot(x='RainTomorrow',data=df)
sns.despine()
```

```
sns.countplot(x='RainToday',data=df)
sns.despine()
```

```
sns.distplot(df['WindSpeed3pm'])
```

```
sns.barplot(x='Cloud9am', y='Temp9am' ,data=df)
sns.despine()
```

```
sns.countplot(x='Cloud9am' ,data=df)
```
